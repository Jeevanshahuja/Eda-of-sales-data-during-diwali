# COUNT OF PURCHASES BY AGE GROUP AND GENDER

```python
age_gender_counts = df.groupby(['Age Group', 'Gender']).size().unstack().fillna(0)
age_groups = age_gender_counts.index.tolist()
male_counts = age_gender_counts['M'].tolist()
female_counts = age_gender_counts['F'].tolist()
x = np.arange(len(age_groups))
width = 0.35
fig, ax = plt.subplots()
bars1 = ax.bar(x - width/2, male_counts, width, label='Male',edgecolor='black',linewidth=3)
bars2 = ax.bar(x + width/2, female_counts, width, label='Female',edgecolor='black',linewidth=3)
ax.set_xlabel('Age Group')
ax.set_ylabel('Count')
ax.set_title('Count of Purchases by Age Group and Gender')
ax.set_xticks(x)
ax.set_xticklabels(age_groups)
ax.legend()
```

```python
amount_age=df.groupby("Age Group").size()
amount_age
```

# AMOUNT SPENT BY EACH AGE GROUP

```python
age_groups = amount_age.index
counts = amount_age.values
plt.figure(figsize=(10, 6))
plt.bar(age_groups, counts, color='blue',edgecolor="black",linewidth=3)
plt.xlabel('Age Group')
plt.ylabel('Amount')
plt.title('Amount by Age Group')
```

```python
marital_count=df.groupby("Marital_Status").size()
marital_count
```

# NUMBER OF ORDERS VS MARITAL STATUS

```python
sns.barplot(x=marital_count.index,y=marital_count.values,edgecolor='black',linewidth=3)
plt.xticks([0,1],["Not Married","Married"])
```

# AMOUNT VS MARITAL STATUS

```python
grouped_marital = df.groupby(['Marital_Status', 'Gender'])['Amount'].sum().reset_index()
sns.barplot(x='Marital_Status', y='Amount', hue='Gender', data=grouped_marital,edgecolor='black',linewidth=3)
plt.xlabel('Marital Status')
plt.ylabel('Amount')
plt.xticks([0,1],["Not Married","Married"])
plt.title('Amount vs Marital Status')
plt.legend(title='Gender')
```
